#### complement

```python
mean = 12.58942199
median = 9.348684
i=0
```

```python
def complement (a, b):
    return random.uniform(min(a, b),max(a, b))
```

```python
for index, row in newscore.iterrows():
    if np.isnan(row['score']):
        print(index)
        i+=1
print(i)
```

```python
for index, row in newscore.iterrows():
    if np.isnan(row['score']):
        newscore.iloc[index,5] = complement(mean, median)
    print(row['score'])
```
